```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_boston
```

```python
boston = load_boston()
boston.keys()
```

```python
data = boston["data"]
data.shape
```

```python
print(boston.DESCR)
```

```python
feature_names = boston["feature_names"]
feature_names
```

```python
target = boston["target"]
```

```python
X = pd.DataFrame(data, columns=boston.feature_names)
X.head()
```

```python
X.info()
```

```python
y = pd.DataFrame(target, columns=["price"])
y.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred.shape
```

```python
check_test = pd.DataFrame({'y_test': y_test['price'], 'y_pred': y_pred.flatten(),})
check_test.head(10)
```

```python
check_test["error"] = check_test["y_pred"] - check_test["y_test"]
check_test.head(10)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score_1 = r2_score(y_test, y_pred)
r2_score_1
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('creditcard.csv')
df.head(10)
```

```python
df["Class"].value_counts(normalize=True)
```

```python
df.info()
df.isnull().astype(np.int64).sum().astype(np.int32)
```

```python
pd.options.display.max_columns = 100
```

```python
df.head(10)
```

```python
X = df.drop("Class", axis=1)
X.head()
```

```python
y = df["Class"]
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state=100, stratify = y)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
parameters = [{"n_estimators": [10, 15],
              'max_features':np.arange(3, 5),
              'max_depth':np.arange(4, 7)}]
```

```python
model2 = GridSearchCV(
    estimator = RandomForestClassifier(random_state=100),
    param_grid=parameters,
    scoring="roc_auc",
    cv=3)
```

```python
model2.fit(X_train, y_train )
```

```python
model2.best_params_
```

```python
pred_proba = model2.predict_proba(X_test)
pred_proba
```

```python
y_pred_proba = y_pred_proba[:, 1]
y_pred_proba
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_pred_proba)
```
